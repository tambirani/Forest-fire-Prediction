```
import xarray as xr

ds = xr.open_dataset('/content/final_fire_weather_dataset.nc')
ds
```

```
import numpy as np

X = np.stack([
    ds['tem_c'].values,
    ds['tp'].values,
    ds['wind_speed'].values,
    ds['d2m'].values
], axis=-1)

y = ds['fire'].values

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```
X_flat = X.reshape(-1, 4)
y_flat = y.reshape(-1)

print(X_flat.shape)
print(y_flat.shape)
```

```
import numpy as np

# fire indices
fire_idx = np.where(y_flat == 1)[0]

# no fire indices
nofire_idx = np.where(y_flat == 0)[0]

# equal no fire select
nofire_sample = np.random.choice(nofire_idx, size=len(fire_idx), replace=False)

# combine
final_idx = np.concatenate([fire_idx, nofire_sample])

X_bal = X_flat[final_idx]
y_bal = y_flat[final_idx]

print(X_bal.shape, y_bal.shape)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_bal, y_bal, test_size=0.2, random_state=42
)

print(X_train.shape, X_test.shape)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# create model
model = LogisticRegression(max_iter=1000)

# train
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)

# accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rf = RandomForestClassifier(
    n_estimators=300,     # trees increase
    max_depth=18,         # deeper learning
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```
import torch
import torch.nn as nn
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_bal)

X_tensor = torch.tensor(X_scaled, dtype=torch.float32)
y_tensor = torch.tensor(y_bal, dtype=torch.long)

print(X_tensor.shape)
```

```
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# dataset
dataset = TensorDataset(X_tensor, y_tensor)
loader = DataLoader(dataset, batch_size=512, shuffle=True)

# transformer model
class FireTransformer(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(4, 64)
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=64, nhead=4),
            num_layers=2
        )
        self.fc2 = nn.Linear(64, 2)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = x.unsqueeze(1)          # add sequence dim
        x = self.transformer(x)
        x = x.squeeze(1)
        return self.fc2(x)

model = FireTransformer()

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# training
for epoch in range(5):
    total_loss = 0
    for xb, yb in loader:
        pred = model(xb)
        loss = criterion(pred, yb)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss:", total_loss)
```

```
# evaluation
model.eval()

with torch.no_grad():
    outputs = model(X_tensor)
    _, predicted = torch.max(outputs, 1)
    correct = (predicted == y_tensor).sum().item()
    accuracy = correct / len(y_tensor)

print("Transformer Accuracy:", accuracy)
```

```
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# reshape for CNN (samples, channels, features)
X_cnn = torch.tensor(X_bal, dtype=torch.float32).unsqueeze(1)
y_cnn = torch.tensor(y_bal, dtype=torch.long)

dataset = TensorDataset(X_cnn, y_cnn)
loader = DataLoader(dataset, batch_size=512, shuffle=True)

# CNN model
class FireCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv1d(1, 16, kernel_size=2)
        self.conv2 = nn.Conv1d(16, 32, kernel_size=2)
        self.fc1 = nn.Linear(32*2, 64)
        self.fc2 = nn.Linear(64, 2)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))
        x = x.view(x.size(0), -1)
        x = self.relu(self.fc1(x))
        return self.fc2(x)

model = FireCNN()

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# training
for epoch in range(5):
    total_loss = 0
    for xb, yb in loader:
        pred = model(xb)
        loss = criterion(pred, yb)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss:", total_loss)
```

```
# evaluation
model.eval()

with torch.no_grad():
    outputs = model(X_cnn)
    _, predicted = torch.max(outputs, 1)
    correct = (predicted == y_cnn).sum().item()
    acc = correct / len(y_cnn)

print("CNN Accuracy:", acc)
```

```
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# tensor data
X_dnn = torch.tensor(X_bal, dtype=torch.float32)
y_dnn = torch.tensor(y_bal, dtype=torch.long)

dataset = TensorDataset(X_dnn, y_dnn)
loader = DataLoader(dataset, batch_size=512, shuffle=True)

# deep neural network
class DNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(4, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 2)
        )

    def forward(self, x):
        return self.net(x)

model = DNN()

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# training
for epoch in range(10):
    total_loss = 0
    for xb, yb in loader:
        pred = model(xb)
        loss = criterion(pred, yb)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss:", total_loss)
```

```
# evaluation
model.eval()

with torch.no_grad():
    outputs = model(X_dnn)
    _, predicted = torch.max(outputs, 1)
    correct = (predicted == y_dnn).sum().item()
    acc = correct / len(y_dnn)

print("DNN Accuracy:", acc)
```
